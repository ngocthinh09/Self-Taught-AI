Dự đoán giá nhà:
- Tạo tập dữ liệu bằng các cặp giá trị (diện tích, giá nhà)
- Huấn luyện mô hình hồi quy tuyến tính xấp xỉ tốt nhất với dữ liệu
- Sử dụng mô hình để dự đoán giá của một căn nhà có diện tích cho trước


```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn as skl
from sklearn import linear_model
from numpy import array
```

```python
# Xây dựng dataset dùng train model
data = array([[15,1504], [17,1608],[75,7394],[10,831],[50,5507],
              [35,3011],[90,8612],[80,9036],[5,614],[65,5832]])

superficies = data[:,0]
price = data[:,1]
```

```python
plt.title('Sự phụ thuộc của giá nhà vào diện tích')
plt.xlabel('Diện tích')
plt.ylabel('Giá nhà')
plt.scatter(superficies,price,color = 'red')
```

```python
# Xây dựng model linear regression
model = linear_model.LinearRegression()
model.fit(superficies.reshape(-1,1),price)
```

```python
# Xây dựng đồ thị
plt.title('Sự phụ thuộc của giá nhà vào diện tích')
plt.xlabel('Diện tích')
plt.ylabel('Giá nhà')
plt.scatter(superficies,price,color = 'red')
plt.plot(superficies,model.predict(superficies.reshape(-1,1)),color = 'blue')
```

```python
# Dự đoán với các giá trị đầu vào
data_predict = array([100, 59, 1999])
for val_predict in data_predict:
  print(model.predict([[val_predict]]))
```
