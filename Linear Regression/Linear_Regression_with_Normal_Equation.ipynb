### Linear Regression
- Sử dụng công thức nghiệm tìm điểm tối ưu của bài toán Linear Regression
- Xây dựng model dự đoán cân nặng của một người thông qua chiều cao

```python
# khai báo các thư viện cần thiết
from __future__ import division
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
```

```python
# height : X_axis    weight : Y_axis
# chuẩn bị dữ liệu
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
Y = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
```

```python
plt.title('Model predicting weight based on height')
plt.axis([140,190,45,75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.plot(X,Y,'ro')
plt.show()
```

```python
# xây dựng ma trận x mở rộng X_bar
one_col = np.ones((X.shape[0],1))
X_bar = np.concatenate((one_col,X),axis = 1)

# tìm nghiệm tối ưu
A = (X_bar.T @ X_bar)
B = (X_bar.T @ Y)
W = (np.linalg.pinv(A) @ B)

print('w =',W)

# dựng đồ thị
w_0 = W[0][0]
w_1 = W[1][0]
x0 = np.linspace(145,185,2)
y0 = w_0 + x0*w_1

plt.title('Model predicting weight based on height')
plt.axis([140,190,45,75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.plot(X,Y,'ro')

plt.plot(x0,y0,color = 'blue')
plt.show()
```

```python
# cho mô hình dự đoán thử
value_pred = [155,160]

for i in range(len(value_pred)):
  print(f'Person {i}: Height: {value_pred[i]} --> Weight: {(w_0 + value_pred[i]*w_1):.2f}')
```
