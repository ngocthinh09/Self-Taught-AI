```python
import numpy as np
import pandas as pd
```

```python
# Đọc và load dữ liệu điểm và lưu vào một dataframe
data = pd.read_excel("test_data/data_ma005.xlsx", index_col = "STT")
data.head()
```

```python
# Điền tất cả những ô điểm trống bằng giá trị 0
data[data.isnull().any(axis = 1)]
data.fillna(value = 0, inplace = True)
data.head()
```

```python
# Tính cột điểm trung bình dựa trên ba cột "Coursework", "Midterm" và "Final" với trọng số tương ứng là 0.2, 0.2, 0.6
data["GPA"] = (0.2*data["Coursework"] + 0.2*data["Midterm"] + 0.6*data["Final"]).round(1)
data.head()
```

Quan sát ngẫu nhiên một sinh viên, tính xác suất sinh viên đó qua học phần.

```python
# Lọc ra các sinh viên có GPA >= 5.0
filter = data[data["GPA"] >= 5.0]
# Đếm số lượng sinh viên qua môn
passed = filter.shape[0]
print(f"Tống số lượng sinh viên: {data.shape[0]}")
print(f"Số lượng sinh viên qua môn: {passed}")
```

Quan sát ngẫu nhiên một sinh viên, thấy sinh viên đó thi chưa đạt kì thi giữa kì, tính xác suất người đó thi đặt kì thi cuối kì

```python
# Lọc ra các sinh viên có cột Midterm < 5.0
not_pass_midterm = data[data["Midterm"] < 5.0]
# Lọc ra các sinh viên có cột Final >= 5.0 từ các sinh viên có cột Midterm < 5.0
pass_final = not_pass_midterm[not_pass_midterm["Final"] >= 5.0]
print(f"Số lượng sinh viên thi chưa đạt giữa kì : {not_pass_midterm.shape[0]}")
print(f"Số lượng sinh viên thi đạt cuối kì nhưng thi không đạta kì : {pass_final.shape[0]}")
pass_final.head()
```

Quan sát ngẫu nghiên một sinh viên thì thấy sinh viên đó thi đạt. Tính xác suất sinh viên này thi chưa đạt ở kỳ thi giũa kỳ

```python
# Lọc từ ra các sinh viên qua môn từ các sinh viên thi chưa đặt giữa kì
pass_ = not_pass_midterm[not_pass_midterm["GPA"] >= 5.0]
print(f"Số lượng sinh viên qua môn mà thi chưa đạt giữa kì: {pass_.shape[0]}")
```
